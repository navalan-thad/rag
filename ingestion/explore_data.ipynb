```python
# SciFact
from beir import util
from beir.datasets.data_loader import GenericDataLoader

url = "https://public.ukp.informatik.tu-darmstadt.de/thakur/BEIR/datasets/scifact.zip"
data_path = util.download_and_unzip(url, "datasets")
corpus, queries, qrels = GenericDataLoader(data_folder=data_path).load(split="test")
```

```python
print(f"Corpus size: {len(corpus)}")
print(f"Queries: {len(queries)}")
print(f"Qrels: {len(qrels)}")

# Log the average document length in tokens, the min/max, and how many queries have 1 vs 2+ relevant docs
from collections import Counter
import numpy as np
doc_lengths = [len(doc["text"].split()) for doc in corpus.values()]
print(f"Avg doc length (tokens): {np.mean(doc_lengths):.2f}")
print(f"Min doc length (tokens): {np.min(doc_lengths)}")
print(f"Max doc length (tokens): {np.max(doc_lengths)}")
relevance_counts = Counter(len(rel_docs) for rel_docs in qrels.values())
print(f"Queries with 1 relevant doc: {relevance_counts[1]}")
print(f"Queries with 2+ relevant docs: {sum(count for rel, count in relevance_counts.items() if rel > 1)}")
```
